**Importing Libraries:**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Loading Grayscale Image for further Process:**

```python
image_path = '/content/drive/MyDrive/DIU/Final Semester/DIP/DIP Assignment/Grayscale_Cat.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

**1. Histogram Equalization + Median Filtering**

```python
# Applying histogram equalization
equalized_image = cv2.equalizeHist(image)

# Applying median filtering
median_filtered_image = cv2.medianBlur(equalized_image, 5)

# Displaying images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Image'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(median_filtered_image, cmap='gray'), plt.title('Filtered Image'), plt.axis('off')
plt.show()
```

**2. CLAHE + Gaussian Filtering**

```python
# Applying CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(image)

# Applying Gaussian filtering
gaussian_filtered_image = cv2.GaussianBlur(clahe_image, (5, 5), 0)

# Displaying images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(clahe_image, cmap='gray'), plt.title('CLAHE Image'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(gaussian_filtered_image, cmap='gray'), plt.title('Filtered Image'), plt.axis('off')
plt.show()
```

**3. Adaptive Histogram Equalization + Unsharp Masking**

```python
# Applying AHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
ahe_image = clahe.apply(image)

# Applying unsharp masking
blurred_image = cv2.GaussianBlur(ahe_image, (5, 5), 0)
unsharp_mask = cv2.addWeighted(ahe_image, 1.5, blurred_image, -0.5, 0)

# Displaying images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(ahe_image, cmap='gray'), plt.title('AHE Image'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(unsharp_mask, cmap='gray'), plt.title('Sharpened Image'), plt.axis('off')
plt.show()
```

**4. Gamma Correction + Bilateral Filtering**

```python
# Applying gamma correction
gamma = 1.5
gamma_corrected_image = np.array(255 * (image / 255) ** gamma, dtype='uint8')

# Applying bilateral filtering
bilateral_filtered_image = cv2.bilateralFilter(gamma_corrected_image, 9, 75, 75)

# Displaying images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(gamma_corrected_image, cmap='gray'), plt.title('Gamma Corrected Image'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(bilateral_filtered_image, cmap='gray'), plt.title('Filtered Image'), plt.axis('off')
plt.show()
```
